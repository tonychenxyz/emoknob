```python
import os

# Define the directory containing the files
directory = "/Users/tonychen.xyz/Documents/文件/Columbia/Research/emo-knob-webpage/audios/neutral_audios"

# Iterate over each file in the directory
for filename in os.listdir(directory):
    # Split the filename to extract the emotion and index
    parts = filename.split('_')
    if "neutraltext" in filename:
        idx = [part for part in parts if "neutraltext" in part][0].split('text')[-1].split('.')[0]
        emotion = parts[0]
    elif len(parts) >= 3:
        emotion = parts[0]
        idx = parts[-1].split('.')[0]  # Extract the index before the file extension
    elif len(parts) == 2:
        emotion = parts[0]
        idx = parts[1].split('.')[0]  # Extract the index before the file extension
    else:
        continue  # Skip files that do not match the expected patterns

    # Construct the new filename
    new_filename = f"{emotion}_{idx}.wav"

    print(filename, new_filename)

    os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))


```
